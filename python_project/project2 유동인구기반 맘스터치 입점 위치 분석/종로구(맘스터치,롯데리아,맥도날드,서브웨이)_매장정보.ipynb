```python
import folium
```

```python
#매장주소 위도, 경도로 변환
address= df['매장주소']
print(address.head())
```

```python
for i in range(len(address)):
    a = address[i].split(' ')
    address[i] = " ".join(a[0:4])
print(address)
```

```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
latitude = []
longitude =[]

for i in address:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
#매장주소 위도, 경도로 변환 후 데이터프레임 완성
df1 = pd.DataFrame({'패스트푸드종류': df['label'],'매장명':df['매장명'],'매장주소':address,'위도':latitude,'경도':longitude})
```

```python
df1
```

```python
#위도,경도 0인 값 삽입
df1.loc[0,'위도'] = 37.585897675
df1.loc[0,'경도'] = 126.995735023561
df1.loc[17,'위도'] = 37.573037774
df1.loc[17,'경도'] = 126.972709184907
```

```python
df1
```

```python
#매장별 위치 색깔로 지도에 표시
colors = {'롯데리아' : 'red', '서브웨이' : 'green','맘스터치':'blue','맥도날드':'orange'}

m = folium.Map(location=[37.57037778, 126.9816417],
               zoom_start=14,
               )
df1.apply(lambda row:folium.CircleMarker(location=[row["위도"], row["경도"]], popup=row["패스트푸드종류"], color=colors[row['패스트푸드종류']]).add_to(m), axis=1)
m
```

```python
#매장별 위치 마크로 표시
m = folium.Map(location=[37.57037778, 126.9816417],
               zoom_start=14,
               )
df1.apply(lambda row:folium.Marker(location=[row["위도"], row["경도"]], popup=row["패스트푸드종류"]).add_to(m), axis=1)
m
```
