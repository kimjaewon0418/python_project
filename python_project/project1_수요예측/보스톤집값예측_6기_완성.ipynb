# Data Preperation

## 후보변수 선택

### 상관계수 도출
* `데이터프레임.corr()`
* 상관계수는 -1부터 1 사이의 값을 가진다.
* 상관계수가 -1에 가까울수록 음(-)의 상관관계가 강하며, +1에 가까울수록 양(+)의 상관관계가 강하다는 의미이다.
* 상관계수를 이용하여 후보변수를 선택한다.
    * 상관계수가 0.7 이상이면 강한 상관관계가 있다고 할 수 있다.
    * 상관계수가 0.3 이상이면 상관관계가 있다고 할 수 있다.
    * 상관계수가 0.1 이상이면 약한 상관관계가 있다고 할 수 있다.
    * 상관계수가 0.1 미만이면 상관관계가 거의 없다고 할 수 있다.

```python
# 독립변수 중 종속변수와 선형관계를 확인하기 위한 변수를 선택한다.
# df.columns : 'CHAS'를 제외하고 모두 선택
# ['CRIM', 'ZN', 'INDUS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX','PTRATIO', 'B', 'LSTAT', 'MEDV']
```

```python
cols = ['CRIM', 'ZN', 'INDUS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX','PTRATIO', 'B', 'LSTAT', 'MEDV']

plt.figure(figsize=(15,5))
sns.heatmap(abs(df[cols].corr()), annot=True, cmap='Blues')

# plt.figure(figsize=(15,5))
# sns.heatmap(abs(df[cols].corr()), annot=True, cmap='Blues')
```

###  다중공선성 분석
* 독립 변수 간에 높은 상관관계가 있을 때 나타난다.
* 다중공선성은 모델을 더 복잡하게 만들 수 있으며, 모델의 예측 성능을 저하시킬 수 있습니다.
* 독립 변수 간 상관관계가 0.7 이상이면 다중공선성을 제거한다.
    * INDUS, NOX
    * INDUS, DIS
    * INDUS, TAX
    * NOX, AGE
    * NOX, DIS
    * DIS, AGE
    * TAX, RAD
* NOX, DIS, TAX는 변수로 선택하지 않는다.

```python
# 다중공선성을 제거한 변수를 이용
# 상관계수의 절댓값이 높은 순으로 정렬
cols = ['CRIM', 'ZN', 'INDUS', 'RM', 'AGE', 'RAD', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
abs(df[cols].corr()['MEDV']).sort_values(ascending=False)
```

    * 상관관계가 높은 변수는 직선 부분에 데이터가 많이 몰려있고 밀도가 촘촘하다. 이 밀도를 표현한 숫자를 '상관계수'라고 부른다.

```python
# 후보변수 : 'LSTAT', 'RM' , 'PTRATIO', 'INDUS' ,'CRIM', 'RAD', 'AGE', 'ZN', 'B', 'CHAR'
# 후보변수 간의 상관관계

sns.pairplot(data=df
            , x_vars=['LSTAT', 'RM' , 'PTRATIO', 'INDUS' ,'CRIM', 'RAD', 'AGE', 'ZN', 'B', 'CHAS']
            , y_vars=['MEDV']
            , kind='reg')
```

## 데이터셋 준비

### 변수 선택
* 전진 선택 (Forward Selection): 변수를 하나씩 추가하면서 모델을 구축하고, 각 단계에서 성능을 평가하여 가장 유용한 변수를 선택합니다.
* 후진 제거 (Backward Elimination): 모든 변수를 포함한 상태에서 하나씩 제거하면서 모델을 구축하고, 각 단계에서 성능을 평가하여 불필요한 변수를 제거합니다.

### 훈련세트, 테스트세트 나누기
* sklearn패키지의 model_selection 모듈에 있는 train_test_split함수 사용
    * train_test_split함수에 독립변수, 종속변수를 전달
    * test_size의 디폴트값은 0.25
    * test_size의 크기를 변경할 수 있음.
    * 독립변수와 종속변수를 각각 훈련세트, 테스트세트로 분리하여 리턴해준다.

```python
# 후보변수 : ['LSTAT', 'RM' , 'PTRATIO', 'INDUS' ,'CRIM', 'RAD', 'AGE', 'ZN', 'B', 'CHAS']
X = df[['LSTAT', 'RM' , 'PTRATIO', 'CHAS']] # 독립변수는 2차원이어야 한다.
y = df['MEDV']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.25)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# 모델 학습

## 선형회귀

```python
# 모델 생성 : 빈 모델 객체를 생성한다.
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()

# 모델 훈련 : 훈련세트로 모델 훈련
lr_model.fit(x_train, y_train)

# 모델의 훈련 결과(기울기, 절편)
print(f'''모델의 훈련 결과
coef:{lr_model.coef_}
intercept:{lr_model.intercept_}
''')

# 모델 평가
# 훈련데이터, 테스트데이터를 이용한 예측값
test_predict = lr_model.predict(x_test)
# print(test_predict)
# print(y_test)
# print(x_test)

# 실제값과 예측값을 이용하여 모델 평가
from sklearn.metrics import r2_score, mean_squared_error
print('R2:',r2_score(y_test, test_predict))
print('MSE:',mean_squared_error(y_test, test_predict))
```

## KNeighborsRegressor

```python
# 모델 생성 : 빈 모델 객체를 생성한다.
from sklearn.neighbors import KNeighborsRegressor

knr_model = KNeighborsRegressor()

# 모델 훈련 : 훈련세트로 모델 훈련
knr_model.fit(x_train, y_train)

# 모델 평가
# 훈련데이터, 테스트데이터를 이용한 예측값
test_predict = knr_model.predict(x_test)

# 실제값과 예측값을 이용하여 모델 평가
from sklearn.metrics import r2_score, mean_squared_error
print('R2:',r2_score(y_test, test_predict))
print('MSE:',mean_squared_error(y_test, test_predict))
```

## DecisionTreeRegressor

```python
# 모델 생성 : 빈 모델 객체를 생성한다.
from sklearn.tree import DecisionTreeRegressor

dtr_model = DecisionTreeRegressor()

# 모델 훈련 : 훈련세트로 모델 훈련
dtr_model.fit(x_train, y_train)

# 훈련 결과
from sklearn.tree import plot_tree
plt.figure(figsize=(10,5))
plot_tree(dtr_model, filled=True, feature_names=cols, fontsize=10,max_depth=2)
# plt.show()

# 모델 평가
# 훈련데이터, 테스트데이터를 이용한 예측값
test_predict = dtr_model.predict(x_test)

# 실제값과 예측값을 이용하여 모델 평가
from sklearn.metrics import r2_score, mean_squared_error
print('R2:',r2_score(y_test, test_predict))
print('MSE:',mean_squared_error(y_test, test_predict))
```

# k-fold 교차검증

## LinearRegressor

```python
# 객체생성
from sklearn.model_selection import KFold
k=5
kf = KFold(n_splits=k, shuffle=True)

# 모델 생성 함수
def fn_model_create(x,y):
    # sklearn의 linear_model 모듈의 LinearRegression 객체 생성
    lr = LinearRegression()
    lr.fit(x, y)
    return lr

# k-fold 교차검증
r2_score_list = []
mse_list = []
for train_idx, test_idx in kf.split(X,y):
    # print('-'*100)
    # print(train_idx, test_idx)
    X_train, y_train = X.iloc[train_idx], y[train_idx]
    X_test, y_test = X.iloc[test_idx], y[test_idx]

    # 훈련 데이터로 학습
    model = fn_model_create(X_train , y_train)

    # 테스트 데이터로 평가하여 리스트에 추가
    test_predict = model.predict(X_test)
    r2_score_list.append(r2_score(y_test, test_predict))
    mse_list.append(mean_squared_error(y_test, test_predict))

print(r2_score_list)
print(mse_list)

print(f'''
r2_score : {sum(r2_score_list)/k}
mse : {sum(mse_list)/k}
''')
```
