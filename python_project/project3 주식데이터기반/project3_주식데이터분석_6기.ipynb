## 업종의 종목 리스트
* 네이버증권
    * **`pd.read_html(url)`** : 해당 url의 테이블 테그 수집

```python
# !pip install beautifulsoup4 --upgrade
# !pip install html5lib --upgrade
```

```python
# 식음료 업종 종목리스트 url
url = 'https://finance.naver.com/sise/sise_group_detail.naver?type=upjong&no=302'


# 웹페이지 URL에서 테이블 읽기
tables = pd.read_html(url, encoding='cp949')
tables
```

```python
tables[2]
```

```python
table = tables[2].dropna(how='all').dropna(axis=1, how='all')
table
```

```python
# 종목명의 '*' 제거
table['Name'] = table['종목명'].str.replace('*','').str.strip()
table['Name']
```

```python
table
```

```python
# fdr 데이터와 병합
df = pd.merge(fdr.StockListing("KRX"), table['Name'])
df
```

```python
detail_info = tables[3].iloc[:,[0,-2]]
detail_info.columns = [0,1]
detail_info = detail_info.set_index(0).T
detail_info['Code']='250000'
detail_info
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import time
detail_info_list = []
for i in df['Code']:
    time.sleep(0.01)
    url = 'https://finance.naver.com/item/main.naver?code='+str(i)    
    tables = pd.read_html(url, encoding='cp949')
    detail_info = tables[3].iloc[:,[0,-2]]    
    detail_info.columns = [0,1]
    detail_info = detail_info.set_index(0).T
    detail_info['Code']=i
    detail_info_list.append(detail_info)
    print(i, end='\t')
```

```python
df_detail = pd.concat(detail_info_list)
df_detail
```

```python
df_detail = pd.concat(detail_info_list)
df_detail = df_detail.dropna(axis=1, how='all').iloc[:,:-3]
df_detail
```

```python
df_all = pd.merge(df, df_detail, on='Code')
df_all
```

### 시각화

```python
df_all['영업이익률'] = df_all['영업이익률'].astype(float)
df_all['PER(배)'] = df_all['PER(배)'].astype(float)
temp = df_all.sort_values('영업이익률', ascending=False).head(20)
temp = temp[['Name','영업이익률','PER(배)','PBR(배)']]
temp
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
sns.barplot(data = temp, y='Name', x='영업이익률')
plt.title('영업이익률')

plt.subplot(1,3,2)
sns.barplot(data = temp, y='Name', x='PER(배)')
plt.title('PER(배)')

plt.subplot(1,3,3)
sns.barplot(data = temp, y='Name', x='PBR(배)')
plt.title('PBR(배)')

plt.tight_layout()
```
